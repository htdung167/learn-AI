### Pyplot API vs Object-Oriented API
* Pyplot API --> Quickly
* OOP API --> Hard plot and Advanced

```python
#Pyplot API
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color = "blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)");
plt.xlabel("x");
plt.ylabel("F(x)");
# plt.xlim([0,8]);
# plt.ylim([-2, 2]);
#Hoac co the thay bang: plt.axis([xmin, xmax, ymin, ymax]);
# plt.axis([0, 4, -0.7, 0.6]);
# plt.axis("tight"); #fit 
plt.axis("equal");
plt.legend(); # de xuat hien chu thich
plt.title("Sin(x) and Cos(x)");
plt.show() #Dung khi run tren terminal
```

```python
#OO API
#1. Prepare data
x = [1,3,6,9]
y = [10, 27, 98, 56]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height
# fig la hinh tong, ax la hinh nho

#3. Plot the data
ax.plot(x, y);

ax.set(title="A simple plot",
      xlabel="x",
      ylabel="y");
```

### Line

```python
x = np.linspace(0, 10, 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**5, label="x**5");
ax.set(title="Test Plot", xlabel="x", ylabel="y");
ax.legend();
```

### Scatter

```python
# Pyplot API
plt.scatter(x, x**5);
```

```python
# OO API
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
size = 500 * rng.rand(100)
color = rng.rand(100)
```

```python
fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s=size, c=color, cmap="viridis", alpha = 0.7);
#alpha dieu chinh do trong suot
#cmap dieu chinh he mau sac
fig.colorbar(img1); #Hien thi thanh colorbar
```

### 3. Bar

### 4. Histogram

```python
#Prepare data
np.random.seed(1)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10]
```

```python
#Pyplot API
plt.hist(student_height, bins=30); #bins: chia nho
```

```python
# OO API
fig, ax = plt.subplots()
ax.hist(student_height, bins=50);
```

### 5. Subplots


```python
#Option1:
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));
# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x/3);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(dic.keys(), dic.values());
ax4.hist(student_height, bins=50);
```

```python
#Option2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5));
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x/3);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(dic.keys(), dic.values());
ax[1,1].hist(student_height, bins=50);
```
