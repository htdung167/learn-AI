## Data Manipulation

```python
df.item_price.dtype
```

### Apply

```python
df.item_price.apply(lambda x: float(x.replace('$', '')))
```

```python
#Same
df["item_price"].apply(lambda x: float(x.replace('$', ''))).dtype
```

```python
df.item_price = df.item_price.apply(lambda x: float(x.replace('$', '')))
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe(include='all')
```

### Add Column

```python
df["total_price"] = df.quantity * df.item_price
```

```python
df.head()
```

```python
#Tổng toàn bộ doanh thu
df.total_price.sum()
```

### Group by

```python
#Món nào được order nhiều nhất
df.groupby("item_name")["quantity"].apply(print)
```

```python
#Tính tổng số lần order theo item_name
c = df.groupby("item_name")["quantity"].sum()
```

```python
c
```

```python
# Tăng dần
c.sort_values()
```

```python
#Giảm dần
c.sort_values(ascending=False).head(5)
```
